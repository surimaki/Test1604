```python
filtered_df = df[df["ข้อมูล"].str.contains("Average Expenditure Per Day - Visitor - Foreigner|Average Expenditure Per Day - Visitor - Thai")]
print(filtered_df)
```

```python
# ลบ , และแปลงเป็น float (เฉพาะคอลัมน์ปี)
cols = df.columns[2:]  # ข้าม "ข้อมูล" และ "จังหวัด"
df[cols] = df[cols].replace(',', '', regex=True).astype(float)
```

```python
for t in ["Foreigner", "Thai", "Total"]:
    sub = df[df["ข้อมูล"].str.contains(f"Average Expenditure Per Day.*{t}")]
    avg = sub[cols].mean().mean()
    print(f"ค่าเฉลี่ยต่อวัน ({t}): {round(avg, 2)} บาท")

```
